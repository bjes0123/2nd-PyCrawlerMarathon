```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
r1=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response1 = r1.text

data = json.loads(response)
data1 = json.loads(response1)
sum=0
for d in data:
    sum+=d['commentCount']
    print(sum)
print(sum/30)

sum=0
for d in data1:
    sum+=d['commentCount']
    print(sum)
print(sum/30)

sum=0
for d in data:
    sum+=d['likeCount']
    print(sum)
print(sum/30)

sum=0
for d in data1:
    sum+=d['likeCount']
    print(sum)
print(sum/30)
```
